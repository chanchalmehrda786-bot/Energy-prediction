```python
import matplotlib.pyplot as plt
```

```python
def draw_flowchart():
    fig, ax = plt.subplots(figsize=(8, 12))
    ax.axis("off")

    steps = [
        "Start",
        "Collect Delhi Energy Data",
        "Data Cleaning & Preprocessing",
        "Exploratory Data Analysis",
        "Feature Selection",
        "Train ML Model",
        "Predict Energy Consumption",
        "Calculate Carbon Emissions",
        "Generate Reports",
        "End"
    ]

    x = 0.5
    y_positions = list(range(len(steps), 0, -1))

 #  THIS IS THE KEY FIX
    ax.set_xlim(0, 1)
    ax.set_ylim(0, len(steps) + 1)

# Draw boxes
    for step, y in zip(steps, y_positions):
        ax.text(
            x, y, step,
            ha="center", va="center",
            fontsize=11,
            bbox=dict(boxstyle="round,pad=0.4", edgecolor="black")
        )
# Draw arrows
    for i in range(len(y_positions) - 1):
        ax.annotate(
            "",
            xy=(x, y_positions[i + 1] + 0.3),
            xytext=(x, y_positions[i] - 0.3),
            arrowprops=dict(arrowstyle="->", linewidth=1.5)
        )

    ax.set_title(
        "Flowchart: Energy Prediction & Carbon Emission Reduction",
        fontsize=14, pad=20
    )

    plt.show()


draw_flowchart()
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(14, 4))
ax.axis('off')

components = [
    ("Data Source\n(Delhi Energy Data)", 0.05),
    ("Preprocessing", 0.28),
    ("ML Prediction\nModel", 0.50),
    ("Carbon Emission\nCalculation", 0.72),
    ("Reports &\nVisualization", 0.92)
]

for text, x in components:
    ax.text(
        x, 0.5, text,
        ha='center', va='center',
        bbox=dict(boxstyle="round,pad=0.4"),
        fontsize=11
    )

ax.set_title("System Architecture Diagram", fontsize=14)
plt.show()
```
